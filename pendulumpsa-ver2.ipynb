```python
import matplotlib.pyplot as plt
import numpy as np

# npoints = 41
npoints = 41
X = np.zeros((4,npoints))
X[0:2,:] = np.loadtxt('/content/drive/MyDrive/Colab Notebooks/pendulumfiles/pendulumdata1.1.txt')[:,:npoints] #These come from the pendulum data.
X[2:4,:] = np.loadtxt('/content/drive/MyDrive/Colab Notebooks/pendulumfiles/pendulumdata2.1.txt')[:,:npoints]
Xavg = np.mean(X,axis=1) # Compute mean
B = X - np.tile(Xavg,(npoints,1)).T # Mean-subtracted data
obs = B.T
U, S, VT = np.linalg.svd(obs,full_matrices=0) # PCA Analysis

from PIL import Image

# This section computes for the PCA of the data.
plt.figure(figsize=(6,10))
plt.subplot(311)
plt.scatter(X[0],-X[1],color="blue",s=4)
#plt.figure(figsize=(8,13.5))
plt.subplot(312)
plt.scatter(X[2],-X[3],color="blue",s=4)
#plt.figure(figsize=(8,13.5))
#plt.subplot(313)
#plt.scatter(X[4],-X[5],color="blue",s=4)

# This portion computes for the singular value and the cumulative energy.
fig, ax = plt.subplots(1,2,figsize=(6,3))
ax[0].set_title("Singular Values $\\sigma_r$")
ax[0].semilogy(S,'-o',color='k')
ax[0].set_xlabel("$r$")
ax[1].set_title("Commulative Energy")
ax[1].set_xlabel("$r$")
ax[1].plot(np.cumsum(S)/np.sum(S),'-o',color='k')
plt.tight_layout()
plt.show()
print(np.cumsum(S)/np.sum(S))

#This portion lists down the PCA of the data.
fig1 = plt.figure(figsize=(8,8))
ax = fig1.add_subplot(111, projection='3d')
x = VT[0,:] @ obs.T
y = VT[1,:] @ obs.T
z = VT[2,:] @ obs.T
ax.scatter(x,z,y,color='k')
ax.set_xlim(-600,600)
ax.set_ylim(-500,500)
ax.set_zlim(-500,500)
ax.set_yticks(np.arange(-500,500,100))
ax.set_zticks(np.arange(-500,500,100))
ax.set_aspect("equal")
ax.plot([-800,800],[0,0],[0,0],color="r")
ax.plot([0,0],[-200,200],[0,0],color="b")
ax.plot([0,0],[0,0],[-200,200],color="g")
ax.text(820,0,-10, "PC1", color="r")
ax.text(0,220,0, "PC3", color="b")
ax.text(-30,0,220, "PC2", color="g")
plt.title("Projection on the First 3 Principal Components")

#This shows the top view of the pendulum.
fig1 = plt.figure(figsize=(8,8))
ax = fig1.add_subplot(111)
x = VT[0,:] @ obs.T
y = VT[1,:] @ obs.T
z = VT[2,:] @ obs.T
ax.scatter(x,y,color='k')
ax.set_xlim(-500,500)
ax.set_ylim(-300,800)
ax.set_yticks(np.arange(-200,200,200))
ax.set_aspect("equal")
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
plt.title("TOP VIEW")

#This shows the 1st of 2 side views.
fig1 = plt.figure(figsize=(2,2))
ax = fig1.add_subplot(111)
x = VT[0,:] @ obs.T
y = VT[1,:] @ obs.T
z = VT[2,:] @ obs.T
ax.scatter(y,z,color='k')
ax.set_xlim(-400,400)
ax.set_ylim(-400,400)
ax.set_yticks(np.arange(-400,400,100))
ax.set_aspect("equal")
ax.set_xlabel("PC3")
ax.set_ylabel("PC2")
plt.title("FRONT/SIDE VIEW")

#This shows the 2nd of 2 side views.
fig1 = plt.figure(figsize=(8,8))
ax = fig1.add_subplot(111)
x = VT[0,:] @ obs.T
y = VT[1,:] @ obs.T
z = VT[2,:] @ obs.T
ax.scatter(x,z,color='k')
ax.set_xlim(-900,900)
ax.set_ylim(-500,500)
ax.set_yticks(np.arange(-500,500,100))
ax.set_aspect("equal")
ax.set_xlabel("PC1")
ax.set_ylabel("PC3")
plt.title("ANOTHER FRONT/SIDE VIEW")
```
